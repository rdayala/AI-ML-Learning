  2. sepal width in cm 
  3. petal length in cm 
  4. petal width in cm 
  5. class: 
     - Iris Setosa

## Worked Example

Let $X$ be the two class IRIS Dataset. Let $A$ be a $2 \times 4$ matrix as given below:

\begin{equation*}
A=
\begin{bmatrix}
   1 & 0 & 0 & 0\\
   0 & 1 & 0 & 0\\
\end{bmatrix}
\end{equation*}

 
 1. Compute $X' = AX$.
 2. Plot this $X'$ on 2D graph with class1 as red and class2 as blue.


```python
#Import required packages
import numpy as np
import matplotlib.pyplot as plt

#Extract features and labels in separate variables
Features = iris_data.iloc[:,0:4].values
Labels = iris_data.iloc[:,4:5].values
print(Features)
print(Labels)
```

```python
## Now store Feature matrix as X
X = Features

## Create matrix A as given in the Exercise above
A = np.array([[1,0,0,0],[0,0,0,1]])
print(A)
```

```python
## As A is 2*4 matrix. To compute X' convert X into 4*150 using transpose function. 
##As for matrix multiplication dimensions should match
X_transpose = np.transpose(X)
print(X_transpose)
```

```python
## Compute X' same as X1. Only the name is changed. X1 is now 2*150 dimensional.
## Meaning we have extracted 2 features out of 4
###################
## X1 = np.matmul(A,X_transpose)
## Python3 has a new symbol for matrix multiplications
###################
X1 = A @ X_transpose
print(X1.shape)
```

```python
## Convert the matrix X1 back to 150*2 and append label creating X1_final of 150*3
X1 = np.transpose(X1)
X1_final = np.hstack((X1,Labels))
print(X1_final)
```

#### Putting it all together

```python
## Now store Feature matrix as X
X = Features

## Create matrix A as given in the Exercise above
A = np.array([[1,0,0,0],[0,0,0,1]])

## As A is 2*4 matrix. To compute X' convert X into 4*150 using transpose function. 
##As for matrix multiplication dimensions should match
X_transpose = np.transpose(X)

## Compute X' same as X1. Only the name is changed. X1 is now 2*150 dimensional.
## Meaning we have extracted 2 features out of 4
X1 = A @ X_transpose

## Convert the matrix X1 back to 150*2 and append label creating X1_final of 150*3
X1 = np.transpose(X1)
X1_final = np.hstack((X1,Labels))

##Plot the points on graph and visualize.
plt.figure(1, figsize=(20,10))
plt.scatter(X1_final[:,0],X1_final[:,1],c=X1_final[:,2],s=60)
plt.show()
```

# Exercise 1
Repeat the above Example with a new matrix A, given below;

\begin{equation*}
A2 =
\begin{bmatrix}

\end{bmatrix}
\end{equation*}

 1. Compute $X^{\prime} = AX$ ?

```python
#Import required packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

### Convert the labels column as 0 for "iris-setosa", 1 for "iris-versicolor" and 2 for "iris-virginica"  
### Store the converted data in variable iris_data
def irisLabel(s):
    s = s.lower()
    if s == "iris-setosa":
        return 0
    if s == "iris-versicolor":
        return 1
    if s == "iris-virginica":
        return 2

iris_data = pd.read_csv("../Datasets/bezdekIris.data",header=None, converters={4:irisLabel})

#Extract features and labels in separate variables
Features = iris_data.iloc[:,0:4].values
Labels = iris_data.iloc[:,4:5].values

## Now store Feature matrix as X
X = Features

## Create matrix A as given in the Exercise above
A2 = np.array([[0,1,0,0],[0,0,1,0]])

## As A is 2*4 matrix. To compute X' convert X into 4*150 using transpose function. 
##As for matrix multiplication dimensions should match
X_transpose = np.transpose(X)

## Compute X' same as X2. Only the name is changed. X2 is now 2*150 dimensional.
## Meaning we have extracted 2 features out of 4
X2 = A2 @ X_transpose

## Convert the matrix X2 back to 150*2 and append label creating X2_final of 150*3
X2 = np.transpose(X2)
X2_final = np.hstack((X2,Labels))

##Plot the points on graph and visualize.
def colorLabelMap(XLabel):
    colorLabels = []
    for index in range(XLabel.shape[0]):
        if XLabel[index] == 0.0:
            colorLabels.append('r')
        if XLabel[index] == 1.0:
            colorLabels.append('b')
        if XLabel[index] == 2.0:
            colorLabels.append('g')
    return colorLabels

##Plot the points on graph and visualize.
plt.figure(1, figsize=(20,10))
plt.scatter(X2_final[:,0],X2_final[:,1],c=colorLabelMap(X1_final[:,2]),s=60)
plt.show()
```
